```python
import numpy as np
import scipy.stats as stats
import pandas as pd

# Elegir el segmento y la variable continua
segmento_elegido = 'alta_gruesa'
calorias_segmento = df[df['segmento'] == segmento_elegido]['calorias (kcal)']

# Verificamos si hay datos nulos
if calorias_segmento.isnull().sum() > 0:
    print("Error: La variable contiene valores nulos.")
else:
    # Parámetros de la distribución normal
    mu, sigma = np.mean(calorias_segmento), np.std(calorias_segmento)

    # Creamos intervalos (bins)
    num_bins = 10  # Ajustar según el análisis
    bins = np.linspace(calorias_segmento.min(), calorias_segmento.max(), num_bins + 1)

    # Frecuencias observadas
    observed, _ = np.histogram(calorias_segmento, bins=bins)

    # Calculamos las frecuencias esperadas bajo la normal
    expected_raw = len(calorias_segmento) * (
        stats.norm.cdf(bins[1:], loc=mu, scale=sigma)
        - stats.norm.cdf(bins[:-1], loc=mu, scale=sigma)
    )

    # Verificamos que las frecuencias esperadas no tengan problemas
    if expected_raw.sum() == 0 or len(expected_raw) != len(observed):
        print("Error: Las frecuencias esperadas no se calcularon correctamente.")
    else:
        # Ajustamos las frecuencias esperadas
        expected = expected_raw * (observed.sum() / expected_raw.sum())

        # Test de chi-cuadrado
        chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

        # Definimos nivel de confianza 95%
        alpha = 0.05

        print(f"Estadístico Chi-cuadrado: {chi2_stat}")
        print(f"Valor p: {p_value}")

        if p_value > alpha:
            print("No se rechaza la hipótesis nula: los datos podrían provenir de una distribución normal.")
        else:
            print("Se rechaza la hipótesis nula: los datos no se ajustan a una distribución normal.")
```

Como podemos observar en los gráficos, el único segmento que sigue una distribución normal es alta_gruesa, lo cual está respaldado por el resultado del test de bondad de ajuste y su representación gráfica en forma de campana simétrica. En contraste, los gráficos de los demás segmentos muestran características que claramente se desvían de la normalidad, como asimetrías, picos adicionales o formas bimodales. Esto nos permite concluir que, visualmente, los demás segmentos no siguen una distribución normal, reafirmando la necesidad de analizar cada segmento de forma individual antes de asumir que se ajustan a un mismo tipo de distribución.

```python
import numpy as np
from scipy import stats

# Parámetros de la muestra S1
media_muestral_S1 = muestra_S1['calorias (kcal)'].astype(float).mean()
desviacion_estandar_S1 = muestra_S1['calorias (kcal)'].astype(float).std()
n_S1 = len(muestra_S1)

# Nivel de significancia
nivel_significancia = 0.05  # para un test de dos colas al 95%

# Valor crítico (distribución normal estándar para dos colas)
valor_critico = stats.norm.ppf(1 - nivel_significancia/2)

# Cálculo de la región de rechazo
error_estandar = desviacion_estandar_S1 / np.sqrt(n_S1)

# Límites de la región de rechazo
limite_inferior = media_muestral_S1 - valor_critico * error_estandar
limite_superior = media_muestral_S1 + valor_critico * error_estandar

print(f"Región de rechazo para S1 (nivel 95%): x ∈ (-∞, {limite_inferior:.4f}] ∪ [{limite_superior:.4f}, +∞)")
```

```python
#Dejar fijos los bordes de la región de rechazo y estimar la curva de operación variando la media y graficar.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

sigma = desviacion_estandar_S1
n = n_S1
media_muestral = media_muestral_S1

z_alpha = stats.norm.ppf(1 - nivel_significancia/2)
x = np.linspace(media_muestral - 5 * sigma, media_muestral + 5 * sigma, 100)
y = stats.norm.pdf(x, loc=media_muestral, scale=sigma)

plt.plot(x, y, label='Distribución Normal')
plt.fill_between(x, y, where=(x < limite_inferior) | (x > limite_superior), color='red', alpha=0.5)
plt.axvline(x=limite_inferior, color='red', linestyle='--', label='Región de Rechazo')
plt.axvline(x=limite_superior, color='red', linestyle='--')
plt.xlabel('Media')
plt.ylabel('Densidad de Probabilidad')
plt.title('Curva de Operación para S1')
plt.legend()
plt.grid(True)
plt.show()
```

Toda la zona roja es la región de rechazo, mientras que la zona central que esta en blanco es la region de no rechazo

```python
#Concluimos si la varaible X es significativa para la variable Y con un 99% de confianza para este segmento
def preparar_datos(df, segmento):
    """
    Prepara los datos para el análisis de regresión.
    """
    segment_data = df[df['segmento'] == segmento]
    X = sm.add_constant(segment_data['calorias (kcal)'])
    Y = segment_data['gramos (g)']
    return X, Y, segment_data

def realizar_regresion(X, Y):
    """
    Realiza la regresión y retorna el modelo ajustado.
    """
    return sm.OLS(Y, X).fit()

def imprimir_resultados(model, nivel_confianza=0.99):
    """
    Imprime los resultados del análisis de manera organizada.
    """
    alfa = 1 - nivel_confianza
    p_valor = model.pvalues['calorias (kcal)']
    t_valor = model.tvalues['calorias (kcal)']

    print("\nRESULTADOS DEL ANÁLISIS DE SIGNIFICANCIA")
    print("=" * 50)
    print(f"Nivel de confianza: {nivel_confianza * 100}%")
    print(f"Valor alfa (α): {alfa}")
    print(f"Valor p: {p_valor:.10f}")
    print(f"Estadístico t: {t_valor:.4f}")
    print(f"R²: {model.rsquared:.4f}")

    print("\nINTERPRETACIÓN")
    print("=" * 50)
    if p_valor < alfa:
        print(f"Las calorías SON significativas (p-valor: {p_valor:.10f} < {alfa})")
        print(f"Podemos afirmar con {nivel_confianza*100}% de confianza que las")
        print("calorías son significativas para explicar los gramos.")
    else:
        print(f"Las calorías NO son significativas (p-valor: {p_valor:.10f} > {alfa})")

def crear_grafico(segment_data, model, X, Y):
    """
    Crea y muestra el gráfico de regresión.
    """
    plt.figure(figsize=(10, 6))

    # Scatter plot de datos originales
    sns.scatterplot(data=segment_data,
                   x='calorias (kcal)',
                   y='gramos (g)',
                   alpha=0.5)

    # Línea de regresión
    plt.plot(segment_data['calorias (kcal)'],
            model.predict(X),
            color='red',
            label='Línea de regresión')

    # Configuración del gráfico
    plt.title('Análisis de Regresión: Calorías vs Gramos\nSegmento: baja_gruesa')
    plt.xlabel('Calorías (kcal)')
    plt.ylabel('Gramos (g)')
    plt.grid(True, alpha=0.3)

    # Agregar información estadística
    stats_text = (f'R² = {model.rsquared:.4f}\n'
                 f'p-valor = {model.pvalues["calorias (kcal)"]:.10f}')
    plt.text(0.05, 0.95, stats_text,
            transform=plt.gca().transAxes,
            bbox=dict(facecolor='white', alpha=0.8))

    plt.legend()
    plt.show()

def main():
    """
    Función principal que ejecuta todo el análisis.
    """
    # Preparar datos
    X, Y, segment_data = preparar_datos(df, 'baja_gruesa')

    # Realizar regresión
    model = realizar_regresion(X, Y)

    # Mostrar resultados
    imprimir_resultados(model)

    # Crear visualización
    crear_grafico(segment_data, model, X, Y)

# Ejecutar el análisis
if __name__ == "__main__":
    main()
```
